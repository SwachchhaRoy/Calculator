```python
from tkinter import *

# getting the digits from the calculator screen
def get_digit(digit):
    current = result_label['text']
    new = current + str(digit)
    result_label.config(text=new)

# defining the Clear button
def clear():
    result_label.config(text='')

# taking 3 variable for storing the typed values and operator to store the operation
first_no = sec_no = third_no = operator = None


# function for operation
def get_operator(op):
    global first_no,operator
    
    first_no = int(result_label['text'])
    operator = op
    result_label.config(text='')


# function for equal sign
def get_result():
    global first_no,sec_no,operator
    sec_no = int(result_label['text'])

    if operator == '+':
        result_label.config(text=str(first_no + sec_no))
    elif operator == '-':
        result_label.config(text=str(first_no - sec_no))
    elif operator == '*':
        result_label.config(text=str(first_no * sec_no))
    elif operator == '/':
        if sec_no == 0:
            result_label.config(text='Error')
        else:
            result_label.config(text=str(round(first_no / sec_no)))


root = Tk()
root.title('Calculator')  # to give a title to the window
root.geometry('315x410')  # window size
root.resizable(0,0)   # doesnt allow to change the window size
root.configure(background='black')  # to set background color

result_label = Label(root,text='',bg='black',fg='white')     # to create an object for text amd its background and text color
result_label.grid(row=0,column=0,columnspan=10,pady=(50,25),sticky='w')  # to set position of text and pady for positioning in y-axis i.e up and down shift
result_label.config(font=('verdana',30,'bold'))  # to set font name, size and style


# creating buttons
btn7 = Button(root,text='7',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(7))
btn7.grid(row=1,column=0)
btn7.config(font=('verdana',25,'bold'))

btn8 = Button(root,text='8',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(8))
btn8.grid(row=1,column=1)
btn8.config(font=('verdana',25,'bold'))

btn9 = Button(root,text='9',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(9))
btn9.grid(row=1,column=2)
btn9.config(font=('verdana',25,'bold'))

btn_add = Button(root,text='+',bg='#00a65a',fg='white',width=3,command = lambda :get_operator('+'))
btn_add.grid(row=1,column=3)
btn_add.config(font=('verdana',25,'bold'))

btn4 = Button(root,text='4',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(4))
btn4.grid(row=2,column=0)
btn4.config(font=('verdana',25,'bold'))

btn5 = Button(root,text='5',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(5))
btn5.grid(row=2,column=1)
btn5.config(font=('verdana',25,'bold'))

btn6 = Button(root,text='6',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(6))
btn6.grid(row=2,column=2)
btn6.config(font=('verdana',25,'bold'))

btn_sub = Button(root,text='-',bg='#00a65a',fg='white',width=3,command = lambda :get_operator('-'))
btn_sub.grid(row=2,column=3)
btn_sub.config(font=('verdana',25,'bold'))

btn1 = Button(root,text='1',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(1))
btn1.grid(row=3,column=0)
btn1.config(font=('verdana',25,'bold'))

btn2 = Button(root,text='2',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(2))
btn2.grid(row=3,column=1)
btn2.config(font=('verdana',25,'bold'))

btn3 = Button(root,text='3',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(3))
btn3.grid(row=3,column=2)
btn3.config(font=('verdana',25,'bold'))

btn_mul = Button(root,text='*',bg='#00a65a',fg='white',width=3,command = lambda :get_operator('*'))
btn_mul.grid(row=3,column=3)
btn_mul.config(font=('verdana',25,'bold'))

btn0 = Button(root,text='0',bg='#00a65a',fg='white',width=3,command= lambda :get_digit(0))
btn0.grid(row=4,column=0)
btn0.config(font=('verdana',25,'bold'))

btn_div = Button(root,text='/',bg='#00a65a',fg='white',width=3,command = lambda :get_operator('/'))
btn_div.grid(row=4,column=1)
btn_div.config(font=('verdana',25,'bold'))

btn_eq = Button(root,text='=',bg='#00a65a',fg='white',width=3, command=get_result)
btn_eq.grid(row=4,column=2)
btn_eq.config(font=('verdana',25,'bold'))

btn_clr = Button(root,text='C',bg='#00a65a',fg='white',width=3 ,command = lambda :clear())
btn_clr.grid(row=4,column=3)
btn_clr.config(font=('verdana',25,'bold'))





root.mainloop()
```
